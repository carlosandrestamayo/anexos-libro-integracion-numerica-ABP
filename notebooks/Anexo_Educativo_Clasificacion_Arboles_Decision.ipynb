# üìí Clasificaci√≥n con √Årboles de Decisi√≥n en Machine Learning

üéØ **Objetivo:** Aprender a entrenar, evaluar e interpretar un modelo de √Årbol de Decisi√≥n para tareas de clasificaci√≥n con sklearn.


## üå≥ Introducci√≥n a los √Årboles de Decisi√≥n

Los **√Årboles de Decisi√≥n** son modelos ampliamente utilizados en Machine Learning supervisado para problemas de clasificaci√≥n y regresi√≥n. Estos modelos funcionan dividiendo recursivamente el espacio de las caracter√≠sticas en regiones m√°s homog√©neas respecto a la variable objetivo, siguiendo una estructura jer√°rquica similar a un √°rbol.

Su principal ventaja radica en la **interpretabilidad** del modelo: es posible visualizar el conjunto de decisiones que lleva a cada predicci√≥n. Adem√°s, no requieren normalizaci√≥n previa de los datos ni suposiciones de distribuci√≥n.

En clasificaci√≥n, los √°rboles asignan una clase a cada hoja del √°rbol, bas√°ndose en medidas de pureza como el **Gini Index** o la **Entrop√≠a**. El entrenamiento busca minimizar la impureza dentro de cada subconjunto resultante en cada divisi√≥n.

### **Caracter√≠sticas Principales:**
- No requieren normalizaci√≥n de datos.
- Interpretables y f√°ciles de visualizar.
- Pueden sobreajustarse si no se controlan par√°metros como la profundidad m√°xima.

# üìò √Årbol de Decisi√≥n para Clasificaci√≥n con Python y Scikit-learn

## Aplicaciones en Educaci√≥n, Ingenier√≠a y Ciencia de Datos

Este notebook muestra de forma did√°ctica c√≥mo implementar un modelo de **√Årbol de Decisi√≥n** para tareas de clasificaci√≥n supervisada en Machine Learning.  
Incluye:

## üîß Librer√≠as y Herramientas Utilizadas

- `pandas`: manipulaci√≥n de datos tabulares
- `matplotlib`, `seaborn`: visualizaci√≥n gr√°fica
- `sklearn`: Machine Learning (√Årboles de decisi√≥n, m√©tricas, datasets)

Estas librer√≠as son fundamentales en proyectos de Machine Learning supervisado.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Cargar el dataset Iris integrado en sklearn
iris = load_iris(as_frame=True)
X = iris.data
y = iris.target
df = iris.frame

# Visualizar las primeras filas del conjunto de datos
df.head()
```

```python
# Gr√°fica para explorar la relaci√≥n entre largo y ancho del p√©talo
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x=df["petal length (cm)"],
    y=df["petal width (cm)"],
    hue=df["target"],
    palette="deep",
    s=70
)
plt.title("Distribuci√≥n de Clases seg√∫n Caracter√≠sticas de P√©talos (Iris Dataset)")
plt.xlabel("Largo del P√©talo (cm)")
plt.ylabel("Ancho del P√©talo (cm)")
plt.legend(title="Clases")
plt.show()
```

üîç Esta gr√°fica nos permite identificar si las clases son separables visualmente en este conjunto de caracter√≠sticas.


```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"üîπ Tama√±o de entrenamiento: {X_train.shape}")
print(f"üîπ Tama√±o de prueba: {X_test.shape}")
```

üìë Dividimos el dataset en 70% para entrenamiento y 30% para prueba, siguiendo buenas pr√°cticas para evaluar la generalizaci√≥n del modelo.


```python
dtree = DecisionTreeClassifier(random_state=42)
dtree.fit(X_train, y_train)
```

üéØ Entrenamos un **√Årbol de Decisi√≥n** con los datos de entrenamiento. El modelo genera reglas de decisi√≥n que permiten clasificar las muestras seg√∫n las caracter√≠sticas.


```python
plt.figure(figsize=(12, 6))
plot_tree(dtree, feature_names=X.columns, class_names=iris.target_names, filled=True)
plt.title("Estructura del √Årbol de Decisi√≥n Entrenado")
plt.show()
```

Visualizamos la estructura del √°rbol para interpretar c√≥mo el modelo toma decisiones basadas en los atributos del conjunto de datos.


```python
y_pred = dtree.predict(X_test)

print("‚úÖ Accuracy:", accuracy_score(y_test, y_pred))
print("\nüìÑ Classification Report:\n", classification_report(y_test, y_pred))
print("\nüî¢ Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

üìä Evaluamos el rendimiento del modelo utilizando:
- **Accuracy**: porcentaje global de aciertos.
- **Classification Report**: m√©tricas por clase (precisi√≥n, recall, F1-score).
- **Matriz de Confusi√≥n**: errores y aciertos por clase.


```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel("Predicci√≥n")
plt.ylabel("Valor Real")
plt.title("Matriz de Confusi√≥n - √Årbol de Decisi√≥n (Iris Dataset)")
plt.show()
```

üé® La matriz de confusi√≥n facilita la comprensi√≥n visual de los errores cometidos y refuerza la interpretaci√≥n de las m√©tricas num√©ricas.
