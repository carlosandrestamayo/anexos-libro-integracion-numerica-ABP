# üìä Ejercicio: √Årea Bajo la Curva $ f(x) = \cos(x) $

En este ejercicio se busca representar gr√°ficamente el **√°rea bajo la curva** de la funci√≥n:
$$ f(x) = \cos(x) $$
en el intervalo de  $x = 0$ a $x = \frac{\pi}{2}$, incluyendo:

‚úÖ √Årea sombreada en rojo.  
‚úÖ L√≠neas verticales en incrementos de  $\frac{\pi}{8}$.  
‚úÖ Curva completa hasta $x = \pi$.  
‚úÖ Flechas en los ejes para representar prolongaci√≥n.  
‚úÖ Etiquetas en notaci√≥n **LaTeX**.

---




### üìù Prompt utilizado en ChatGPT para generar el c√≥digo de este ejercicio:

> "Genera c√≥digo en Python usando Matplotlib para graficar la funci√≥n coseno desde 0 hasta pi, con √°rea sombreada en rojo desde 0 hasta pi/2, incluyendo l√≠neas verticales en incrementos de pi/8, flechas en los ejes para simular coordenadas prolongadas y etiquetas en LaTeX."

Este prompt fue formulado para obtener un c√≥digo limpio, claro y adecuado para materiales educativos de an√°lisis num√©rico.


````python
import numpy as np
import matplotlib.pyplot as plt

# Datos para la curva completa
x = np.linspace(0, np.pi, 300)
y = np.cos(x)

fig, ax = plt.subplots(figsize=(6, 4))

# Curva completa
ax.plot(x, y, color='blue', linewidth=2, label=r'$f(x) = \cos(x)$')

# √Årea sombreada de 0 a pi/2
x_fill = np.linspace(0, np.pi/2, 200)
y_fill = np.cos(x_fill)
ax.fill_between(x_fill, y_fill, color='red', alpha=0.3)

# L√≠neas verticales en incrementos de pi/8
x_points = np.arange(0, np.pi/2 + 0.01, np.pi/8)
for x0 in x_points:
    y0 = np.cos(x0)
    ax.plot([x0, x0], [0, y0], linestyle='--', color='black', linewidth=1)

# Flechas para ejes personalizados
xmax, ymax = np.pi + 0.2, 1.1
ax.annotate('', xy=(xmax, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))
ax.annotate('', xy=(0, ymax), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))

# Configuraci√≥n de etiquetas
ax.set_xticks([0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi])
ax.set_xticklabels([r'$0$', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$'])
ax.set_yticks([0, 0.5, 1])

ax.set_xlabel(r'$x$', fontsize=12, loc='right')
ax.set_ylabel(r'$f(x)$', fontsize=12, loc='top')

ax.set_xlim(0, xmax)
ax.set_ylim(0, ymax)
ax.grid(False)

# Eliminar todos los spines (bordes del recuadro)
for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)

plt.legend()

# Guardar sin m√°rgenes al exportar
#plt.savefig('coseno_area_sombreada.png', dpi=300, bbox_inches='tight', transparent=True)

plt.tight_layout()
plt.show()
````

## üìä Ejercicio: Representaci√≥n Gr√°fica del √Årea Bajo la Curva $ f(x) = \frac{1}{\sqrt{x^2 + 4}} $

Este ejercicio tiene como prop√≥sito representar gr√°ficamente el √°rea bajo la curva de la funci√≥n
$$f(x) = \frac{1}{\sqrt{x^2 + 4}}$$
en el intervalo de $x = 0$ a $x = 2$, incluyendo:

- El √°rea sombreada en rojo.
- L√≠neas verticales en incrementos de  $0.25$.
- La curva completa hasta $x = 2.5$.
- Flechas en ambos ejes para representar prolongaci√≥n de los ejes coordenados.
- Etiquetas matem√°ticas utilizando notaci√≥n **LaTeX**.


### üìù Prompt utilizado en ChatGPT para generar el c√≥digo de este ejercicio:

> "Genera c√≥digo Python usando Matplotlib para graficar la funci√≥n
f(x) = \frac{1}{\sqrt{x^2 + 4}}  desde 0 hasta 2.5 , con √°rea sombreada en rojo desde 0  hasta 2, incluyendo l√≠neas verticales en incrementos de 0.25 , flechas en los ejes para simular los ejes coordenados prolongados, y etiquetas matem√°ticas en LaTeX."

Este prompt fue formulado para obtener un c√≥digo limpio, claro y adecuado para materiales educativos de an√°lisis num√©rico.


````python
import numpy as np
import matplotlib.pyplot as plt
import os

# Datos para la curva completa
x = np.linspace(0, 2.5, 300)
y = 1 / np.sqrt(x**2 + 4)

fig, ax = plt.subplots(figsize=(6, 4))

# Curva completa
ax.plot(x, y, color='blue', linewidth=2, label=r'$f(x) = \frac{1}{\sqrt{x^2 + 4}}$')

# √Årea sombreada de 0 a 2
x_fill = np.linspace(0, 2, 300)
y_fill = 1 / np.sqrt(x_fill**2 + 4)
ax.fill_between(x_fill, y_fill, color='red', alpha=0.3)

# L√≠neas verticales en incrementos de 0.25
x_points = np.arange(0, 2.01, 0.25)
for x0 in x_points:
    y0 = 1 / np.sqrt(x0**2 + 4)
    ax.plot([x0, x0], [0, y0], linestyle='--', color='black', linewidth=1)

# Ejes personalizados con flechas
xmax, ymax = 2.8, 0.6
ax.annotate('', xy=(xmax, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))
ax.annotate('', xy=(0, ymax), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))

# Configuraci√≥n de etiquetas
ax.set_xticks(np.arange(0, 2.01, 0.25))
ax.set_yticks([])

ax.set_xlabel(r'$x$', fontsize=12, loc='right')
ax.set_ylabel(r'$f(x)$', fontsize=12, loc='top')
ax.set_xlim(0, xmax)
ax.set_ylim(0, ymax)
ax.grid(False)

# ‚úÖ Eliminar el cuadro exterior (spines)
for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
````

````python
# Funci√≥n ajustada para un cambio m√°s brusco en la mitad
x = np.linspace(0, 3, 500)
y = -0.7 * (x - 1.5)**3 + 1.5  # Aument√© el coeficiente y baj√© la altura para hacerlo m√°s visible

# Definir puntos a y b
a, b = 0.7, 2.3
x_fill = np.array([a, b])
y_fill = -0.7 * (x_fill - 1.5)**3 + 1.5

# Crear la figura
fig, ax = plt.subplots(figsize=(6, 4))
fig.patch.set_facecolor('white')

# √Årea sombreada estilo trapecio
ax.fill_between([a, b], [y_fill[0], y_fill[1]], color='lightgrey', alpha=0.5, hatch='///')

# Curva completa
ax.plot(x, y, color='deepskyblue', linewidth=2)

# Puntos a y b destacados
ax.scatter([a, b], y_fill, color='deepskyblue', zorder=5)

# Eliminar bordes (spines)
for spine in ['top', 'right', 'bottom', 'left']:
    ax.spines[spine].set_visible(False)

# Ejes con flechas
ax.annotate('', xy=(3.2, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', color='black'))
ax.annotate('', xy=(0, 2.5), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', color='black'))

# Etiquetas a y b
ax.text(a, -0.2, r'$a$', ha='center', fontsize=12)
ax.text(b, -0.2, r'$b$', ha='center', fontsize=12)

# Etiquetas de ejes
ax.set_xlabel(r'$x$', fontsize=12, loc='right')
ax.set_ylabel(r'$f(x)$', fontsize=12, loc='top')

# L√≠mites y limpieza
ax.set_xlim(0, 3.2)
ax.set_ylim(-0.5, 2.5)
ax.set_xticks([])
ax.set_yticks([])
ax.set_facecolor('white')
ax.grid(False)

plt.tight_layout()
plt.show()
````

````python

# Datos para la funci√≥n sin(x) entre 0 y pi
x = np.linspace(0, np.pi, 300)
y = np.sin(x)

# Crear la figura y ejes
fig, ax = plt.subplots(figsize=(6, 4))

# Graficar la curva sin(x)
ax.plot(x, y, color='blue', linewidth=2, label=r'$f(x) = \sin(x)$')

# Sombrear el √°rea bajo la curva entre 0 y pi
ax.fill_between(x, y, color='red', alpha=0.3)

# L√≠neas verticales en x = 0, pi/6, pi/3, ..., pi
x_points = np.arange(0, np.pi + 0.01, np.pi / 6)
for x0 in x_points:
    ax.plot([x0, x0], [0, np.sin(x0)], linestyle='--', color='black', linewidth=1)

# Eliminar bordes superiores y derechos
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# Posicionar ejes en el origen
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# A√±adir flechas a los ejes
xmax, ymax = 4, 1.2
ax.annotate('', xy=(xmax, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))
ax.annotate('', xy=(0, ymax), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))

# Configurar xticks con fracciones de pi en pasos de pi/6
xticks = [0, np.pi/6, np.pi/3, np.pi/2, 2*np.pi/3, 5*np.pi/6, np.pi]
xticklabels = [r'$0$', r'$\frac{\pi}{6}$', r'$\frac{\pi}{3}$', r'$\frac{\pi}{2}$',
               r'$\frac{2\pi}{3}$', r'$\frac{5\pi}{6}$', r'$\pi$']
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels)

# Configurar yticks
ax.set_yticks([1])

# Cambiar el tama√±o de los ticks
ax.tick_params(axis='both', which='both', labelsize=12)  # Cambia el n√∫mero a lo que necesites

# Etiquetas y l√≠mites
ax.set_xlabel(r'$x$', fontsize=12, loc='right')
ax.set_ylabel(r'$f(x)$', fontsize=12, loc='top')
ax.set_xlim([0, xmax])
ax.set_ylim([0, ymax])

# Quitar cuadr√≠cula
ax.grid(False)

# Mostrar la figura
plt.show()
````

## üíæ ¬øC√≥mo guardar una imagen `.png` en Matplotlib dentro de una carpeta?

Cuando generamos una gr√°fica en **Matplotlib**, podemos guardar la imagen dentro de una carpeta espec√≠fica para organizar mejor nuestros archivos.

### üìå Pasos:

1Ô∏è‚É£ Crea la carpeta (si no existe) utilizando Python.  
2Ô∏è‚É£ Guarda la imagen dentro de esa carpeta con la funci√≥n `plt.savefig()`.

---

### üêç **Ejemplo completo: Guardar en una carpeta llamada `imagenes`**



````python
import os
import numpy as np
import matplotlib.pyplot as plt

# Crear carpeta 'imagenes' si no existe
folder = 'imagenes'
os.makedirs(folder, exist_ok=True)

# Datos para la gr√°fica
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear la gr√°fica
plt.plot(x, y)
plt.title('Ejemplo de gr√°fico guardado en carpeta')

# Definir la ruta completa del archivo
filename = os.path.join(folder, 'grafico_seno.png')

# Guardar la imagen como archivo .png dentro de la carpeta
plt.savefig(filename, dpi=300, bbox_inches='tight')

# Mostrar la gr√°fica
plt.show()
````

### üìÇ ¬øD√≥nde se guarda el archivo?

El archivo **se guarda dentro de la carpeta `imagenes` que se crea en el mismo directorio donde ejecutas tu c√≥digo**.  
En **Google Colab**, esta carpeta aparecer√° en el **explorador de archivos que se encuentra en la parte izquierda de la pantalla**.  
Desde all√≠ podr√°s visualizarla, abrirla o descargar los archivos generados directamente.
